```python
class SmartLight:
    def __init__(self, room_name):
        self.room_name = room_name
        self.status = "OFF"

    def turn_on(self):
        self.status = "ON"
        print(f"{self.room_name} light turned ON")

    def turn_off(self):
        self.status = "OFF"
        print(f"{self.room_name} light turned OFF")

    def display_status(self):
        print(f"{self.room_name} light is {self.status}")

living_room_light = SmartLight("Living room")
bedroom_light = SmartLight("Bedroom")
living_room_light.turn_on()
bedroom_light.turn_off()
living_room_light.display_status()
bedroom_light.display_status()
```

```python
class BankAccount:
    def __init__(self, acc_num, init_bal=0):
        self.acc_num=acc_num
        self.__balance=init_bal
        print(f"Account {self.acc_num} creatrd with initial balance: ${self.__balance:,.2f}")

    def deposit(self, amount):
        if amount>0:
            self.__balance += amount
            print(f"Deposited ${amount:,.2f} into account {self.acc_num}. New balance: ${self.__balance:,.2f}")
        else:
            print("Amount must be positive")

    def withdraw(self, amount):
        if amount > 0:
            if self.__balance>=amount:
                self.__balance-=amount
                print(f"Withdrew ${amount:,.2f} from account {self.acc_num}. New balance: ${self.__balance:,.2f}")
            else:
                print(f"Insufficient funds for withdrawal from account {self.acc_num}. Current balance: ${self.__balance:,.2f}")
        else:
            print("Withdrawal amount must be positive.")

    def get_bal(self):
        return self.__balance

    def display_info(self):
        print(f"\n--- Account Information ({self.acc_num}) ---")
        print(f"Account Number: {self.acc_num}")
        print(f"Current Balance: ${self.get_bal():,.2f}")

account1 = BankAccount("1001", 500)
account2 = BankAccount("1002", 1200)

print("\nTransactions for Account 1001")
account1.deposit(200)
account1.withdraw(100)
account1.withdraw(700)
account1.display_info()

print("\nTransactions for Account 1002")
account2.deposit(300)
account2.withdraw(500)
account2.display_info()
```

```python
class Staff:
    def __init__(self, name, staff_id, department):
        self.name = name
        self.staff_id = staff_id
        self.department = department

    def display_info(self):
        print(f"Name: {self.name}")
        print(f"Staff ID: {self.staff_id}")
        print(f"Department: {self.department}")

class Teacher(Staff):
    def __init__(self, name, staff_id, department, courses, salary):
        super().__init__(name, staff_id, department)
        self.courses = courses
        self.salary = salary

    def teach_course(self, course_name):
        if course_name in self.courses:
            print(f"{self.name} is teaching {course_name}")
        else:
            print(f"{self.name} does not teach {course_name}")

    def display_info(self):
        super().display_info()
        print(f"Type: Teacher")
        print(f"Courses: {', '.join(self.courses)}")
        print(f"Salary: {self.salary} PKR")

class AdministrativeStaff(Staff):
    def __init__(self, name, staff_id, department, role, working_hours):
        super().__init__(name, staff_id, department)
        self.role = role
        self.working_hours = working_hours

    def perform_task(self, task_name):
        print(f"{self.name} ({self.role}) is performing the task: {task_name}")

    def display_info(self):
        super().display_info()
        print(f"Type: Admin Staff")
        print(f"Role: {self.role}")
        print(f"Working Hours: {self.working_hours} hours/week")

class ResearchAssistant(Staff):
    def __init__(self, name, staff_id, department, research_topic, stipend):
        super().__init__(name, staff_id, department)
        self.research_topic = research_topic
        self.stipend = stipend

    def work_research(self):
        print(f"{self.name} is working on research for topic: {self.research_topic}")

    def display_info(self):
        super().display_info()
        print(f"Type: Research Assistant")
        print(f"Research Topic: {self.research_topic}")
        print(f"Stipend: {self.stipend} PKR")

teacher1 = Teacher("Mr. Farooq Zaidi", "CS101", "Computer Science", ["Data Structures", "Artificial Intelligence"], 275000)
admin1 = AdministrativeStaff("Mr. Fahad Samad", "A201", "Admissions", "Registrar", 40)
ra1 = ResearchAssistant("Mr. Ali Jan", "M303", "Computer Engineering", "IT", 25000)

print("\nTeacher Activities")
teacher1.display_info()
teacher1.teach_course("Knowledge Representation & Reasoning")
teacher1.teach_course("Artificial Intelligence")
print("\nAdmin Activities")
admin1.display_info()
admin1.perform_task("process student registrations")
print("\nResearch Assistant Activities")
ra1.display_info()
ra1.work_research()
```

```python
class Vehicle:
    def __init__(self, brand):
        self.brand = brand
        print(f"Vehicle brand initialized: {self.brand}")

class Car(Vehicle):
    def __init__(self, brand, model):
        super().__init__(brand)
        self.model = model
        print(f"Car model initialized: {self.model}")

class ElectricCar(Car):
    def __init__(self, brand, model, battery_capacity):
        super().__init__(brand, model)
        self.battery_capacity = battery_capacity
        print(f"ElectricCar battery capacity initialized: {self.battery_capacity} kWh")

my_electric_car = ElectricCar("Tesla", "Model 3", 75)
print('\n')
print("Car Details")
print(f"Brand: {my_electric_car.brand}")
print(f"Model: {my_electric_car.model}")
print(f"Battery Capacity: {my_electric_car.battery_capacity} kWh")
```

```python
class Notification:
    def send_message(self):
        print("Sending generic notification...")

class EmailNotification(Notification):
    def send_message(self):
        print("Sending Email to user@example.com")

class SMSNotification(Notification):
    def send_message(self):
        print("Sending SMS to +923001234567")

email_notifier = EmailNotification()
sms_notifier = SMSNotification()
email_notifier.send_message()
sms_notifier.send_message()
```

```python
class StudentGrades:
    def __init__(self, student_name):
        self.student_name = student_name
        self.__marks = []
        print(f"Student {self.student_name} initialized")

    def add_marks(self, score):
        if 0 <= score <= 100:
            self.__marks.append(score)
            print(f"{self.student_name}'s score ({score}) has been added")
        else:
            print(f"Invalid Score: {score} (must be between 0 - 100)")

    def get_marks(self):
        return self.__marks

    def display_info(self):
        print(f"\nStudent: {self.student_name}")
        print(f"Marks: {self.get_marks()}")

student1 = StudentGrades("Sahil")
# this will create a new public attribute, not modify private one
student1.__marks = [90, 95] # This creates a NEW public attribute called __marks
print(f"Attempting to set: {student1.__marks}")
# Check private attribute using public getter
print(f"Attempting to get: {student1.get_marks()}")
student1.add_marks(85)
student1.add_marks(72)
student1.add_marks(105) # Invalid
student1.add_marks(-5) # Invalid
student1.add_marks(90)
student1.display_info()
student2 = StudentGrades("Mudassir")
student2.add_marks(60)
student2.add_marks(80)
student2.display_info()
```
