*Task 1:*

```python
import heapq

def beamSearch(graph, start, goal, beam_width):
    beam = [(0, [start])]

    while beam:
        candidates = []
        for cost, path in beam:
            current = path[-1]
            if current == goal:
                return path, cost
            for neighbor, edge_cost in graph.get(current, []):
                new_cost = cost + edge_cost
                new_path = path + [neighbor]
                candidates.append((new_cost, new_path))
        if not candidates:
            break
        beam = heapq.nsmallest(beam_width, candidates, key = lambda x: x[0])

    return None, float('inf')

graph = {
    'S': [('A', 2), ('B', 5), ('C', 4)],
    'A': [('D', 7), ('E', 3)],
    'B': [('F', 6)],
    'C': [('G', 2)],
    'D': [('T', 4)],
    'E': [('T', 6)],
    'F': [('T', 5)],
    'G': [('T', 3)],
    'T': []
}

start = 'S'
goal = 'T'

for k in [1, 2, 3]:
    path, cost = beamSearch(graph, start, goal, k)
    if path:
        print(f"Beam Width = {k}")
        print(f"Path found: {' > '.join(path)}")
        print(f"Total Cost: {cost}")
        print("\n")
    else:
        print(f"Beam Width = {k}: No path found")
```

*Task 2:*

```python
import random

def f(x):
    return -x**2 + 10*x + 5

def hillClimbing():
    curr_x = random.randint(0, 100)
    curr_val = f(curr_x)

    while True:
        neighbors = []

        # left
        if curr_x - 1 >= 0:
            neighbors.append(curr_x - 1)

        # right
        if curr_x + 1 <= 100:
            neighbors.append(curr_x + 1)

        # best
        next_x = curr_x
        next_val = curr_val

        for x in neighbors:
            value = f(x)
            if value > next_val:
                next_x = x
                next_val = value

        # stop if no improv
        if next_val <= curr_val:
            break

        curr_x = next_x
        curr_val = next_val

    return curr_x, curr_val

for i in range(5): # ran this 5 times
    sol_x, sol_val = hillClimbing()
    print(f"Run {i+1}: x = {sol_x}, f(x) = {sol_val}")
```

```python
def calculate_cost(chromosome):
    total = 0
    for task in range(num_tasks):
        machine = chromosome[task]
        total += cost_matrix[task][machine]
    return total

def calculate_fitness(chromosome):
    cost = calculate_cost(chromosome)
    return 1 / cost

def create_random_individual():
    return [random.randint(0, num_machines - 1) for _ in range(num_tasks)]

def select_parents(population):
    pop_sorted = sorted(population, key=lambda x: calculate_fitness(x), reverse=True)
    return pop_sorted[:len(population)//2]

def crossover(parent1, parent2):
    point = random.randint(1, num_tasks - 2)
    child = parent1[:point] + parent2[point:]
    return child

def mutate(chromosome):
    task = random.randint(0, num_tasks - 1)
    chromosome[task] = random.randint(0, num_machines - 1)
    return chromosome

def GA(pop_size):
    population = [create_random_individual() for _ in range(pop_size)]
    
    best_overall = None
    best_fit = 0
    best_gen = 0

    for generation in range(generations):
        fitness_scores = [calculate_fitness(ind) for ind in population]
        
        current_best = max(population, key=calculate_fitness)
        curr_best_fit = calculate_fitness(current_best)

        if curr_best_fit > best_fit:
            best_fit = curr_best_fit
            best_overall = current_best
            best_gen = generation

        parents = select_parents(population)

        new_pop = []
        while len(new_pop) < pop_size:
            p1, p2 = random.sample(parents, 2)
            child = crossover(p1, p2)

            if random.random() < mutation_rate:
                child = mutate(child)

            new_pop.append(child)

        population = new_pop

    best_cost = calculate_cost(best_overall)
    return best_overall, best_cost, best_fit, best_gen

import random

cost_matrix = [
[4, 6, 8, 7, 5],
[7, 5, 6, 8, 4],
[6, 4, 7, 5, 8],
[5, 8, 6, 4, 7],
[8, 6, 5, 7, 4],
[7, 4, 8, 6, 5],
[6, 7, 4, 5, 8],
[5, 6, 7, 8, 4],
[4, 7, 5, 6, 8],
[8, 5, 6, 4, 7]
]

num_tasks = 10
num_machines = 5
mutation_rate = 0.1
generations = 100

for pop_size in [10, 30]:
    solution, cost, fitness, generation = GA(pop_size)
    print(f"\nPopulation Size: {pop_size}")
    print("Best Chromosome:", solution)
    print("Total Cost:", cost)
    print("Fitness:", fitness)
    print("Found at Generation:", generation)
```
