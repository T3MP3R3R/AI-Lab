```python
import random

class Environment:
    def __init__(self):
        self.state = random.choice(['Dirty', 'Clean'])

    def get_percept(self):
        return self.state

    def clean_room(self):
        self.state = 'Clean'

    def random_change(self):
        self.state = random.choice(['Dirty', 'Clean'])


class SimpleReflexAgent:
    def __init__(self):
        pass

    def act(self, percept):
        if percept == 'Dirty':
            return 'Clean the room'
        else:
            return 'Room is already clean'


def run_agent(agent, environment, steps):
    for step in range(steps):
        percept = environment.get_percept()
        action = agent.act(percept)
        print(f"Step {step + 1}: Percept - {percept}, Action - {action}")

        if percept == 'Dirty':
            environment.clean_room()
        environment.random_change()


agent = SimpleReflexAgent()
environment = Environment()
run_agent(agent, environment, 5)
```

```python
class Environment:
    def __init__(self):
        self.grid = {
            'a': 'safe', 'b': 'safe', 'c': 'fire',
            'd': 'safe', 'e': 'fire', 'f': 'safe',
            'g': 'safe', 'h': 'safe', 'j': 'fire'
        }

    def display(self):
        symbols = {
            'fire': 'F',
            'safe': ' '
        }

        print("\nEnvironment Status:")
        print(f"[{symbols[self.grid['a']]}] [{symbols[self.grid['b']]}] [{symbols[self.grid['c']]}]")
        print(f"[{symbols[self.grid['d']]}] [{symbols[self.grid['e']]}] [{symbols[self.grid['f']]}]")
        print(f"[{symbols[self.grid['g']]}] [{symbols[self.grid['h']]}] [{symbols[self.grid['j']]}]\n")

    def has_fire(self, room):
        return self.grid[room] == 'fire'

    def extinguish(self, room):
        self.grid[room] = 'safe'

class Agent:
    def __init__(self, start_room):
        self.position = start_room

    def move_to(self, room):
        self.position = room
        print(f"Robot moved to room {room}")

    def act(self, environment):
        if environment.has_fire(self.position):
            print(f"Fire detected")
            environment.extinguish(self.position)
        else:
            print(f"Room {self.position} is safe")

path = ['a','b','c','d','e','f', 'g', 'h', 'j']
env = Environment()
robot = Agent('a')

print("Initial Environment:")
env.display()
for room in path:
    robot.move_to(room)
    robot.act(env)
    env.display()
print("Final Environment:")
env.display()
```

```python
class Environment:
    def __init__(self):
        self.beds = ['Moist']*9
        self.beds[1] = 'Dry'
        self.beds[4] = 'Dry'
        self.beds[6] = 'Dry'

    def get_status(self, index):
        return self.beds[index]

    def water_bed(self, index):
        self.beds[index] = 'Moist'

    def display_status(self):
        print("\nFinal Bed Status:\n")
        for i, status in enumerate(self.beds):
            print(f"Bed {i + 1}: {status}")

class Agent:
    def __init__(self):
        pass

    def act(self, environment, index):
        number = index+1
        status = environment.get_status(index)

        if status == 'Dry':
            environment.water_bed(index)
            print(f"Robot at Bed {number} is dry, changing to moist")
        else:
            print(f"Robot at Bed {number} is moist, moving to next bed")

environment = Environment()
robot = Agent()

print("Initial Bed Status:\n")
for i, status in enumerate(environment.beds):
    print(f"Bed {i+1}: {status}")

print("\nRobot Actions:\n")
for i in range(len(environment.beds)):
    robot.act(environment, i)
environment.display_status()
```

```python
class Room:
    def __init__(self, name, is_vault=False, has_keycard=False):
        self.name = name
        self.is_vault = is_vault
        self.has_keycard = has_keycard

    def __str__(self):
        return self.name

class Agent:
    def __init__(self):
        self.current_room = None
        self.has_keycard = False

    def move_to(self, room):
        self.current_room = room
        print(f"Agent moved to Room {self.current_room.name}")
        if self.current_room.name == 'B' and self.current_room.has_keycard and not self.has_keycard:
            self.has_keycard = True
            print(f"Keycard acquired")

    def attempt_enter_vault(self, vault_room):
        if vault_room.is_vault:
            if self.has_keycard:
                print(f"Access granted")
            else:
                print(f"Access denied: Keycard required")
        else:
            print(f"Room {vault_room.name} isn't a vault, so no special access required")

room_A = Room('A')
room_B = Room('B', has_keycard=True)
room_C = Room('C')
room_D = Room('D', is_vault=True)

rooms = {'A': room_A, 'B': room_B, 'C': room_C, 'D': room_D}
patrol_path_names = ['A', 'C', 'D', 'B', 'D']
patrol_path = [rooms[name] for name in patrol_path_names]
robot = Agent()

print(f"Initial state:")
print(f"Robot has keycard? {robot.has_keycard}")

for i, next_room in enumerate(patrol_path):
    print(f"\nStep {i + 1}")
    robot.move_to(next_room)
    if next_room.is_vault:
        robot.attempt_enter_vault(next_room)
print("\nPatrol complete")

print(f"\nFinal state:")
print(f"Robot has keycard? {robot.has_keycard}")
```

```python
class Environment:
    def __init__(self, houses):
        self.houses = houses

class Agent:
    def __init__(self, goal):
        self.goal = goal

    def search(self, environment):
        for house in environment.houses:
            print("Visiting:", house)
            if house == self.goal:
                print("Target found: Mission successful")
                return
        print("Target not found")

houses = ['Blue House', 'Green House', 'Red House', 'Yellow House', 'White House']
env = Environment(houses)
drone = Agent("Red House")
drone.search(env)
```

```python
class Environment:
    def __init__(self, name, value, cost):
        self.name = name
        self.value = value
        self.cost = cost

    def utility(self):
        return (self.value * 2) - self.cost


class Agent:
    def choose_best_rock(self, rocks):
        best_rock = None
        best_utility = float('-inf')

        for Environment in rocks:
            u = Environment.utility()
            print(Environment.name, "utility =", u)
            if u > best_utility:
                best_utility = u
                best_rock = Environment

        print("Best choice:", best_rock.name)
rocks = [
    Environment("Rock A", value=5, cost=2),
    Environment("Rock B", value=9, cost=8),
    Environment("Rock C", value=6, cost=3)
]
agent = Agent()
agent.choose_best_rock(rocks)
```

```python
import random

class Environment:
    def __init__(self):
        self.track = {
            'Start': 0,
            'Empty': -1,
            'Cheese': 10,
            'Trap': -10
        }
        self.possible_move_spots = ['Empty', 'Cheese', 'Trap']

    def get_reward(self, spot_name):
        return self.track.get(spot_name, 0)

class Agent:
    def __init__(self):
        self.current_spot = 'Start'

    def make_random_move(self, environment):
        next_spot = random.choice(environment.possible_move_spots)
        self.current_spot = next_spot
        return next_spot

    def receive_feedback(self, environment, spot_name):
        reward = environment.get_reward(spot_name)
        print(f"Mouse moved to: '{spot_name}'")
        print(f"Reward received: {reward} points")

environment = Environment()
mouse = Agent()
print(f"Initial State: Mouse is at '{mouse.current_spot}'\n")
next_spot = mouse.make_random_move(environment)
mouse.receive_feedback(environment, next_spot)
```
